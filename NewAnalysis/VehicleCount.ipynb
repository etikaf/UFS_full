```r
data = read.csv("FullData_TV.csv") # read the data
names(data)
dim(data)
```

## 3:: Separate test and train data using uniform distribution

```r
#data <- data %>% filter(check!=0) # Remove rows with no vehicles
data %>% distinct(ZID) %>% nrow() # find the number of rows select(ZID)

groups <- data %>% select(ZID) %>% distinct(ZID) %>% mutate(prb = runif(1030, 0, 1)) # 1024 is the unique rows

data <- data %>%  left_join(groups) # Join the data with main file
#head(data)

p=0.80 #Split probability

traindata <- filter(data, prb <p) 
testdata <- filter(data, prb >= p) 
dim(traindata)
dim(testdata)
head(testdata)
```

```r
## write to csv
write.csv(traindata, "TrainData_VT.csv", row.names=FALSE)
write.csv(testdata, "TestData_VT.csv", row.names=FALSE)
```
